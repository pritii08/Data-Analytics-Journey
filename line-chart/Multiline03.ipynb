```python
import matplotlib.pyplot as plt

plt.title("Two lines with different width")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

x=[10,20,30]
y1=[20,40,10]
y2=[40,30,10]

plt.plot(x,y1,linewidth=3,color='red',linestyle="--",marker='o')
plt.plot(x,y2,linewidth=5,color='green',linestyle="-.")

plt.show()
```
