```python
import matplotlib.pyplot as plt

plt.title("Two line in same plot with legends")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

x=[10,15,20,25,30]
y1=[20,28,40,20,10]
y2=[40,28,20,28,30]

plt.plot(x,y1,color='blue',label='Line 1')
plt.plot(x,y2,color='green',label='Line 2')

plt.legend()
plt.show()
```
