```python
import matplotlib.pyplot as plt

xdata=["3 Oct","4 Oct","5 Oct","6 Oct","7 Oct"]
y1=[774,776,779,779,779]#open
y2=[776,778,782,780,779]#high
y3=[769,772,775,775,770]#low
y4=[772,776,776,776,775]#close

plt.title("Multiline Plot")
plt.xlabel("Date")
plt.ylabel("Data")

plt.plot(xdata,y1,marker='o',color='red',label="Open")

plt.plot(xdata,y2,marker='s',color='green',label="High")

plt.plot(xdata,y3,marker='^',color='blue',label="Low")

plt.plot(xdata,y4,marker='o',color='yellow',label='Close')

plt.legend()
plt.show()
```
