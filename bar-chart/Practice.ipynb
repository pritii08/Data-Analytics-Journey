```python
"""
Write a Python programming to display a bar chart of the popularity of programming Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
"""

import matplotlib.pyplot as plt

plt.title("Popularity of Programming language Worldwide, Oct 2017")
plt.xlabel("Languages")
plt.ylabel("Popularity")

x=["Java","Python","PHP","Javascript","C#","C++"]
y=[22.2,17.6,8.8,8.7,7.7,6.7]

plt.bar(x,y)

plt.show()
```

```python
"""
Write a Python programming to display a bar chart of the popularity of programming Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
"""

import matplotlib.pyplot as plt

plt.title("Popularity of Programming language\n"+"Worldwide, Oct 2017 compared to year ago")
plt.xlabel("Languages")
plt.ylabel("Popularity")

x=["Java","Python","PHP","Javascript","C#","C++"]
y=[22.2,17.6,8.8,8.7,7.7,6.7]

plt.bar(x,y)

plt.minorticks_on()
plt.grid(which='major',color='red',linestyle='--');
plt.grid(which='minor',linestyle=':',color='black')

plt.show()
```
